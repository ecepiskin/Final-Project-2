```python
df.columns = ['SEMA_M', 'Steaming Hrs', 'REPORT_TYPE', 'Average_Speed', 'Slip', 'VLFO ROB', 'VLSFO Cons.', 'MGO ROB', 'MGO Cons.', 'FW ROB', 'FW Cons.', 'M/E System Lub Oil ROB', 'M/E System Lub Oil Cons' , 'M/E Cylinder Oil ROB', 'M/E Cylinder Oil Cons' ,'D/G Lub Oil ROB', 'D/G Lub Oil Cons', 'Hyd Oil ROB' , 'Hyd Oil Cons' , 'Engine RPM' , 'Engine Load','TC RPM']
```

```python
import pandas as pd

# Convert 'Engine RPM' to numeric (if it's not already)
df['Engine RPM'] = pd.to_numeric(df['Engine RPM'], errors='coerce')  # 'coerce' will turn invalid parsing into NaN

# Filter data based on 'Engine RPM' ranges and calculate the average of 'M/E Cylinder Oil Cons'
def calculate_avg_cylinder_oil(df):
    # RPM Range 70-75
    rpm_70_75 = df[(df['Engine RPM'] >= 70) & (df['Engine RPM'] < 75)]
    avg_70_75 = rpm_70_75['M/E Cylinder Oil Cons'].mean()

    # RPM Range 75-80
    rpm_75_80 = df[(df['Engine RPM'] >= 75) & (df['Engine RPM'] < 80)]
    avg_75_80 = rpm_75_80['M/E Cylinder Oil Cons'].mean()

    # RPM Range 80-85
    rpm_80_85 = df[(df['Engine RPM'] >= 80) & (df['Engine RPM'] < 85)]
    avg_80_85 = rpm_80_85['M/E Cylinder Oil Cons'].mean()

    # Returning results as a dictionary
    return {
        'Average M/E Cylinder Oil Cons (RPM 70-75)': avg_70_75,
        'Average M/E Cylinder Oil Cons (RPM 75-80)': avg_75_80,
        'Average M/E Cylinder Oil Cons (RPM 80-85)': avg_80_85
    }

# Example of how to use this function
result = calculate_avg_cylinder_oil(df)
print(result)
```
